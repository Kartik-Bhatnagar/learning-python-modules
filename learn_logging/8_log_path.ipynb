### Displaying Variable data

```python
import logging
import os
import time
import random
```

```python
cur_dir_path = os.getcwd() #in jupyter notebook we dont have os.path.abspath(__file__) as there is no __file__ variable
level_change ="data/logs"
log_dir_path = os.path.abspath(os.path.join(cur_dir_path,level_change))
if os.path.exists(log_dir_path):
    print(f"Log directory path is valid {log_dir_path}")
else:
    print(f"path doesn't exist {log_dir_path}")
    os.makedirs(log_dir_path)
    print("Created path")
```

```python
file_name = str(int(time.time()))+".log"
file_name
```

```python
file_path = os.path.abspath(os.path.join(log_dir_path,file_name))
```

we will print 10 random numbers between the range of 1 to 1000 with a interval of random sleep between 2 to 10 seconds

first we will configure the logger and then execute our program

```python
logging.basicConfig(filename =file_path,
                    level=logging.INFO,
                    format="{asctime} | {levelname} | :: {message}",
                    style="{",
                    datefmt="%Y-%m-%d %H:%M:%S")
```

```python
sleep_time_low,sleep_time_high = 2,10
```

```python
for i in range(10):
    rand_sleep_time = random.randint(sleep_time_low,sleep_time_high)
    time.sleep(rand_sleep_time)
    rnum = random.randint(1,1000)
    logging.info(f"Random number generated : {rnum} Slept for {rand_sleep_time} seconds ")
    logging.warning("Check if this printed")
```
