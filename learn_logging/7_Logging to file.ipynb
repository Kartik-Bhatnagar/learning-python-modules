### Logging to file

To save your logs in a file, you can set up your loggerâ€™s basicConfig() with the filename argument. Just like when working with files in Python and using the open() function, you must provide a filepath.

```python
import logging
```

```python
logging.basicConfig(filename="logs_view.txt",
                    level=logging.INFO,
                    format = "{asctime}  | {levelname} |-> {message}",
                    style="{",
                    datefmt="%d-%m-%Y %H:%M:%S")
```

```python
logging.info("The code reched here")
```

```python
logging.warning("Oops there is some warning")
```

```python
logging.critical("THere we have a critical issue to look")
```
